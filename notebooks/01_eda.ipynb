```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/raw/creditcard.csv')
print(df.shape)
df.head()
```

```python
df.info()
print("\nMissing values:")
print(df.isnull().sum())
```

```python
fraud_count = df['Class'].value_counts()
print(fraud_count)
print(f"\nFraud percentage: {fraud_count[1]/len(df)*100:.4f}%")

sns.countplot(x='Class', data=df)
plt.title('Fraud vs Normal Transactions')
plt.xticks([0, 1], ['Normal', 'Fraud'])
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 4))

df[df['Class'] == 0]['Amount'].plot(ax=axes[0], kind='hist', bins=50, title='Normal - Amount Distribution')
df[df['Class'] == 1]['Amount'].plot(ax=axes[1], kind='hist', bins=50, title='Fraud - Amount Distribution', color='red')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
corr = df.corr()
sns.heatmap(corr[['Class']].sort_values('Class', ascending=False), 
            annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Feature Correlation with Fraud (Class)')
plt.tight_layout()
plt.show()
```
